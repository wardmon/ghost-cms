# PANDAS IO - Input/output tools

The Pandas i/o API a set of top level reader function accessed like pd.read_csv() that generally return a pandas object. input output 

read_csv_ to_csv

read_table to_table

read_html to_html read_excel to_excel HDF5 format to store file in Hard disk

The two workhorse functions for reading text files (or the flat files) are read_csv and read_table(). They both use the same parsing code to intelligently
convert tabular into a DataFrame object-

read_csv(filepath_or_buffer,sep',', delimiter=None, header='infer, names=None, index_col=None)
         
filepath_or_buffer:str, path object, or file-like object sep : str, default',' Delimiter to use. if sep is None delimiter:str,default None Alias for sep.
        
header:int,list or int, default 'infer' Row number(s) to use as the column names, and the start of the data. Default behavior is to infer the column names
         
names:array-like, optional List of column namesto use.

# Web API -Aplication Programming Interface

API has a set of function that allow data scientists to access specific features or data of an application or other services. API works over the web which can
be accessed using HTTP protocol.

When we want to receive data from an API, we we need to make a requests are used all over the web.

Consider the process of logging into an app such as Medium. you arte given the option to login using google or facebook. In this instance, the Medium app is 
getting access to your google or Facebook user data through an API developed by those companies. APIs are also helpful because they are a straightforword
